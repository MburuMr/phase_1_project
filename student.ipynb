```python
#importing the libraries needed

import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#This code reads and puts the data into a data frame.
aviation_data_df = pd.read_csv('data\Aviation_Data.csv', encoding='latin1', low_memory=False)
aviation_data_df
```

```python
# Shape of the dataset - Displaying the number of rows and columns
print( aviation_data_df.shape)
```

```python
#This provides a concise summary of the dataframe
aviation_data_df.info()
```

```python
# Checking Missing Values 
aviation_data_df.isnull().sum()
```

```python
#Dropping duplicates
aviation_data_df = aviation_data_df.drop_duplicates()
```

```python
aviation_data_df.columns
```

```python
# Standardize column names: lowercase and replace spaces with underscores
aviation_data_df.columns = aviation_data_df.columns.str.lower().str.replace('.', '_')
```

```python
aviation_data_df.columns
```

```python
# Dropping Columns we wont be needing for analysis 
aviation_data_df=aviation_data_df.drop(['schedule', 'air_carrier' ,'far_description', 'longitude', 'latitude', 'airport_code', 'airport_name'], axis=1)
aviation_data_df.head()
```

```python
# Convert event_date column to datetime format
aviation_data_df['event_date'] = pd.to_datetime(aviation_data_df['event_date'])
```

```python
# Extracting the year from the event_date column and creating a new column "Year"
aviation_data_df['year'] = aviation_data_df['event_date'].dt.year
```

```python
#To check some descriptive including the categorical features
aviation_data_df.describe(include='all')
```

```python
#Grouping while summarizing the counts per group 
yr_count=aviation_data_df.groupby('year')['event_id'].count().reset_index()

#visualizing the output
plt.figure(figsize=(10,10))
sns.lineplot(data=yr_count, x='year', y='event_id', marker='o',markersize=15).set(title="Acidents since 1950", xlabel='Period', ylabel=' Number of accidents')
plt.show()
```

```python
# Getting the top 10 aircraft makes
preffered_makes = aviation_data_df['make'].value_counts().head(10)

# Plot the data
plt.figure(figsize=(10, 10))
preffered_makes.plot(kind='bar', color='red', edgecolor='black')



# Add title and labels
plt.title('Top 10 Aircraft Makes with Most Accidents', fontsize=16, fontweight='bold')
plt.xlabel('Aircraft Make', fontsize=12, fontweight='bold')
plt.ylabel('Number of Accidents', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.show()
```

```python
#combine unk and UNK together
aviation_data_df['weather_condition']=aviation_data_df['weather_condition'].replace('Unk','UNK')
```

```python
#Trying to figure out number of accidents based on weather conditions
Weather_condition=aviation_data_df.groupby(['weather_condition','aircraft_damage'])['event_id'].count().reset_index()
plt.figure(figsize=(10,10))
sns.barplot(data=Weather_condition, x='weather_condition', y='event_id', hue='aircraft_damage').set(title="Number of acidents based on weather condition", xlabel='Weather condition & extent of damage', ylabel=' Number of accidents')
plt.show()
```

```python
#Check purposes with accidents

purpose=aviation_data_df.groupby(['purpose_of_flight'])['event_id'].count().reset_index().sort_values(by='event_id', ascending=False).head(10)
plt.figure(figsize=(10,10))
sns.barplot(data=purpose, y='purpose_of_flight', x='event_id').set(title='Number of acidents based on the purpose', ylabel='Purpose of the aircraft', xlabel=' number of accidents')
plt.show()
```

```python
#further checking the accidents occurence with the intended purpose and Engine type of the aircraft

purpose_engt=aviation_data_df.groupby(['purpose_of_flight','engine_type'])['event_id'].count().reset_index().sort_values(by='event_id', ascending=False).head(10)
plt.figure(figsize=(10,10))
sns.barplot(data=purpose_engt, y='purpose_of_flight', x='event_id',hue='engine_type').set(title='Number of acidents based on the purpose and Engine type', ylabel='Purpose of the aircraft', xlabel=' number of accidents')
plt.show()
```

```python
#checking influence of number of engines and the number of accidents

engine=aviation_data_df.groupby(['number_of_engines'])['event_id'].count().reset_index().sort_values(by='event_id', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(data=engine, x='number_of_engines', y='event_id').set(title="Number of acidents based on the number of Engines", ylabel='number of accidents', xlabel=' number of accidents')
plt.show()

```

```python
#combining all the injuries 
aviation_data_df['total_injuries']=aviation_data_df['total_fatal_injuries']+aviation_data_df['total_serious_injuries']+aviation_data_df['total_minor_injuries']
```

```python
#Group total injuries per year
injuries_per_year=aviation_data_df.groupby(['year'])['total_injuries'].sum().reset_index().sort_values(by='year', ascending=True)
```

```python
#Group total of uninjured per year
uninjured_per_year=aviation_data_df.groupby(['year'])['total_uninjured'].count().reset_index().sort_values(by='year', ascending=True)
```

```python
#Plotting 2 graphs showing the changes of injuries and uninjured over time
plt.figure(figsize=(10,10))

sns.lineplot(data=injuries_per_year, x='year', y='total_injuries', label='Total Injuries', marker='s', markersize=7, linestyle='dotted')
sns.lineplot(data=uninjured_per_year, x='year', y='total_uninjured', label='Total Uninjured', marker='D', markersize=5, linestyle='dashdot')

plt.xlabel('Years')
plt.ylabel('Value')
plt.title('Total of injured and uninjured over the years')
plt.legend()
plt.show()
```
